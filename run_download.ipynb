```python
import json
import argparse

def merge_jsonl_files(file1_path, file2_path, output_path):
    """
    Merge two JSONL files into a single output JSONL file.
    
    Args:
        file1_path (str): Path to the first JSONL file
        file2_path (str): Path to the second JSONL file
        output_path (str): Path where the merged JSONL file will be saved
    """
    print(f"Merging {file1_path} and {file2_path} into {output_path}")
    
    # Read and store lines from both files
    merged_lines = []
    
    # Process first file
    with open(file1_path, 'r', encoding='utf-8') as file1:
        for line in file1:
            if line.strip():  # Skip empty lines
                merged_lines.append(line.strip())
    
    # Process second file
    with open(file2_path, 'r', encoding='utf-8') as file2:
        for line in file2:
            if line.strip():  # Skip empty lines
                merged_lines.append(line.strip())
    
    # Write merged content to output file
    with open(output_path, 'w', encoding='utf-8') as output_file:
        for line in merged_lines:
            output_file.write(line + '\n')
    
    print(f"Successfully merged {len(merged_lines)} JSON lines")

merge_jsonl_files('articles_2023/all_articles.jsonl',
                  'articles_2024/all_articles.jsonl',
                  'all_articles.jsonl')
```
